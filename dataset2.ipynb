```python
#K-means Clustering Code
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate some random data
np.random.seed(42)
X = np.random.rand(100, 2)

# Create a KMeans instance with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the KMeans instance to the data
kmeans.fit(X)

# Predict the cluster for each data point
labels = kmeans.predict(X)
print(labels)
# Plot the data points with cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', marker='*')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
#Hierarichal Clustering
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
# Sample data
np.random.seed(42)
data = np.random.rand(10, 2)  # 10 samples, 2 features
# Perform hierarchical/agglomerative clustering
Z = linkage(data, 'ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, labels=[f'Point {i}' for i in range(1, len(data) + 1)])
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
#Hierarichal Clustering Dendogram with different linkage
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Sample data
np.random.seed(42)
data = np.random.rand(10, 2)  # 10 samples, 2 features

# Different linkage methods
linkage_methods = ['single', 'complete', 'average']

# Create a subplot for each linkage method
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, method in zip(axes, linkage_methods):
    # Perform hierarchical clustering
    Z = linkage(data, method)

    # Plot the dendrogram
    dendrogram(Z, labels=[f'Point {i}' for i in range(1, len(data) + 1)], ax=ax)
    ax.set_title(f'Hierarchical Clustering Dendrogram ({method.capitalize()} linkage)')
    ax.set_xlabel('Sample Index')
    ax.set_ylabel('Distance')

plt.tight_layout()
plt.show()
```

```python
#Agglomerative Clustering
import numpy as np
from sklearn.cluster import AgglomerativeClustering

# Sample data
np.random.seed(42)
data = np.random.rand(10, 2)  # 10 samples, 2 features

# Different linkage methods
linkage_methods = ['ward', 'complete', 'average']

# Dictionary to store clustering results
clustering_results = {}

for method in linkage_methods:
    # Perform hierarchical clustering
    model = AgglomerativeClustering(linkage=method, n_clusters=3)
    labels = model.fit_predict(data)

    # Store the results
    clustering_results[method] = labels

# Print clustering results
for method, labels in clustering_results.items():
    print(f'Linkage method:', method)
    print(f'Cluster labels:' ,labels)
```
